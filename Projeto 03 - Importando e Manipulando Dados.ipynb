#### Fazendo tratamento na coluna PREÇO UNITÁRIO

```python
base["PREÇO UNITÁRIO"] = base["PREÇO UNITÁRIO"].str.replace(",",".")
```

```python
base["PREÇO UNITÁRIO"] = pd.to_numeric(base["PREÇO UNITÁRIO"],errors="coerce")
```

#### ANÁLISES
- Estoque Atual
- Quantidade de Entradas
- Quantidade de Saídas
- Quantidade de Devoluções
- Estoque por Loja
- Estoque por Categoria
- Estoque por Loja e Categoria

```python
# Estoque Atual
display(base["MOVIMENTAÇÃO"].sum())
```

```python
# Estoque por loja
display(base.groupby("ID LOJA")["MOVIMENTAÇÃO"].sum(numeric_only=True))
```

```python
# Estoque por categoria
display(base.groupby("CATEGORIA")["MOVIMENTAÇÃO"].sum(numeric_only=True))
```

```python
# Estoque por loja e categoria
display(base.groupby(["ID LOJA","CATEGORIA"])["MOVIMENTAÇÃO"].sum(numeric_only=True))
```

```python
display(base.groupby("TIPO")["MOVIMENTAÇÃO"].sum(numeric_only=True))
```

```python
# Quantidade de Devoluções
display(base[base['TIPO'] == "DEVOLUÇÃO"].groupby("TIPO")["MOVIMENTAÇÃO"].sum(numeric_only=True))
```

```python
# Quantidade de Entradas
display(base[base['TIPO'] == "ENTRADA"].groupby("TIPO")["MOVIMENTAÇÃO"].sum(numeric_only=True))
```

```python
# Quantida de Saídas
display(base[base['TIPO'] == "SAÍDA"].groupby("TIPO")["MOVIMENTAÇÃO"].sum(numeric_only=True))
```

##### Conhecendo os recursos pela Análise de Estoque por Loja

```python
# !pip install plotly_express
import plotly_express as px
```

```python
display(base)
```

```python
px.histogram(base,x="ID LOJA",y="MOVIMENTAÇÃO",text_auto=True,color="ID LOJA")
```

```python
# Gráfico de Estoque por Categoria
gec = px.histogram(base,x="CATEGORIA",y="MOVIMENTAÇÃO",text_auto=True,color="CATEGORIA")
```

```python
# Gráfico de Estoque por Loja e Categoria
gelc = px.histogram(base,x=["CATEGORIA","ID LOJA"],y="MOVIMENTAÇÃO",text_auto=True,color="CATEGORIA")
gelc.write_html("Gráfico de Estoque por Loja e Categoria.html")
```

```python
# Gráfico de Quantidade de Entradas, Saídas e Devoluções
px.histogram(base,x="TIPO",y="MOVIMENTAÇÃO",text_auto=True,color="TIPO")
```

```python
px.histogram(base,x=["TIPO","ID LOJA"],y="MOVIMENTAÇÃO",text_auto=True,color="TIPO")
```

```python
gec.write_html("Gráfico Estoque por Categoria.html")
```
